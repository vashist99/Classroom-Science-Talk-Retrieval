```python
import pandas as pd
import os
from dotenv import load_dotenv
import requests

load_dotenv()

api_key = os.getenv("LLM_API_KEY")
model_tracka = os.getenv("LLM_MODEL_TRACKA")
model_trackb = os.getenv("LLM_MODEL_TRACKB")
model_augment = os.getenv("LLM_MODEL_AUGMENT")
completion_url = os.getenv("COMPLETION_URL")
embedding_url = os.getenv("EMBEDDING_URL")
```

```python
def connect_to_llm(api_key):
    headers = {"Authorization": f"Bearer {api_key}"}
    body1 = {
        "model": model_augment,
        "messages": [
            {
                "role": "user",
                "content": "Write a 2 line poem about a cat."
            }
        ],
        "temperature": 0.7,
        "max_tokens": 100,
        "top_p": 1,
        "frequency_penalty": 0,
        "presence_penalty": 0
    }
    body2 = {
        "model": model_trackb,
        "messages": [
            {
                "role": "user",
                "content": "Write a 2 line poem about a cat."
            }
        ],
        "temperature": 0.7,
        "max_tokens": 100,
        "top_p": 1,
        "frequency_penalty": 0,
        "presence_penalty": 0
    }
    response1 = requests.post(f"{completion_url}", headers=headers, json=body1)
    response2 = requests.post(f"{completion_url}", headers=headers, json=body2)

    return response1.json(), response2.json()
```

```python
res1, res2 = connect_to_llm(api_key)
print(res1)
print(res2)
```
